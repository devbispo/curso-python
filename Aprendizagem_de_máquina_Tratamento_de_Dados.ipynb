```python
#substituindo as idades com problema pela média das idades corretas 
import pandas as pd
import numpy as np
base = pd.read_csv('credit_data.csv')
base.mean()
print('Média:', base['age'][base.age > 0].mean()) #Média total das idades 

base.loc[base.age < 0, 'age'] = base['age'][base.age>0].mean() 
base.loc[base['age']<0]
print(base.loc[[15]])
print(base.loc[[21]])
print(base.loc[[26]])
```

```python
from numpy.ma.core import mean
import pandas as pd
import numpy as np
base = pd.read_csv('credit_data.csv')
previsores = base.iloc[:, 1:4].values
classe = base.iloc[:, 4].values
print(np.argwhere(np.isnan(previsores)))#posicào de valores faltantes
print(previsores[np.argwhere(np.isnan(previsores))[0][0],:])
print(previsores[np.argwhere(np.isnan(previsores))[1][0],:])
print(previsores[np.argwhere(np.isnan(previsores))[2][0],:])
from sklearn.impute import SimpleImputer # subtituir nulos pela média
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(previsores[:,0:3])
previsores[:, 0:3] = imputer.transform (previsores [:, 0:3])
print(np.argwhere(np.isnan(previsores)))
print (previsores[28,:])
```
